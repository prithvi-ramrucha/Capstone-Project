### **Generating Dataset**

This notebook was used to generate the query dataset which details all known and generated queries for each function of the BBO capstone challenge.

```python
# Dependencies,
import numpy as np
import pandas as pd
```

Let us write a function to return a Pandas `DataFrame` whose rows are the evaluations in the BBO challenge. Known evaluations supplied for the challenge possess negative integers in `Query Number` field. On the other hand, those generated by the optimiser are associated with positive integers in the same field. 

```python
def create_df(func_num):

    # Reading Excel file for optimiser evalutions,
    df = pd.read_excel(f"Function {func_num}/evals.xlsx", header=1)
    df = df.loc[:, ~df.columns.str.contains("Unnamed")]

    # Loading known evalations (warm-start evals),
    X, y = np.load(f"Function {func_num}/initial_inputs.npy"), np.load(f"Function {func_num}/initial_outputs.npy")

    # Creating rows and concatening with existing df,
    rows = []
    for k, (x, fx) in enumerate(zip(X, y), start=1):
        row = {"Query Number": -k}
        row.update({
            f"x{i+1}": val for i, val in enumerate(x)
        })
        row["f(X)"] = fx
        rows.append(row)

    new_rows = pd.DataFrame(rows)
    df_new = pd.concat([new_rows, df], ignore_index=True)
    
    return df_new
```

Now we create a data frame for each black-box function and save it as a CSV file,

```python
# Creating dataframe,
j = 8
df = create_df(func_num=j)

# Saving to CSV,
df.to_csv(f"dataset/func_{j}_evals.csv", index=False)
df
```
