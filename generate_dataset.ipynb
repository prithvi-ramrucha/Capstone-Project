### **Generating Dataset**

```python
# Dependencies,
import numpy as np
import pandas as pd
```

Let us write a function to return a Pandas `DataFrame` whose rows are evaluations in the BBO challenge. Known evaluations provided for the challenge possess a negative query numbers while those generated by the optimiser are associated with position query numbers. 

```python
def create_df(func_num):

    # Reading Excel file for optimiser evalutions,
    df = pd.read_excel(f"Function {func_num}/evals.xlsx", header=1)
    df = df.loc[:, ~df.columns.str.contains("Unnamed")]

    # Loading known evalations (warm-start evals),
    X, y = np.load(f"Function {func_num}/initial_inputs.npy"), np.load(f"Function {func_num}/initial_outputs.npy")

    # Creating rows and concatening with existing df,
    rows = []
    for k, (x, fx) in enumerate(zip(X, y), start=1):
        row = {"Query Number": -k}
        row.update({
            f"x{i+1}": val for i, val in enumerate(x)
        })
        row["f(X)"] = fx
        rows.append(row)

    new_rows = pd.DataFrame(rows)
    df_new = pd.concat([new_rows, df], ignore_index=True)
    
    return df_new
```

Now we create a dataframe for each black-box function and save it as a dataset,

```python
df = create_df(func_num=5)
df
```
